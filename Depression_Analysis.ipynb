**Outlier handling and Data Cleaning (from 2790 rows to 2786 rows)**

```python
import pandas as pd

# Detect outliers and count them
def detect_outliers_with_values(df):
    outlier_info = {}  # To store outlier information
    outlier_values = {}  # To store the specific outlier values

    # Only process columns with data types int64 and float64
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

    for column in numeric_columns:
        Q1 = df[column].quantile(0.25)  # First quartile
        Q3 = df[column].quantile(0.75)  # Third quartile
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_count = outliers.shape[0]

        # If outliers exist, record their information
        if outlier_count > 0:
            outlier_info[column] = outlier_count
            outlier_values[column] = outliers[column].values.tolist()

    return outlier_info, outlier_values

# Detect outliers
outlier_info, outlier_values = detect_outliers_with_values(sampled_df_cleaned)

# Print outlier information and specific values
print("Features with Outliers and their Counts:")
for feature, count in outlier_info.items():
    print(f"{feature}: {count} outliers")

print("\nOutlier Values:")
for feature, values in outlier_values.items():
    print(f"{feature}: {values}")

# Remove rows with outliers
numeric_columns = sampled_df_cleaned.select_dtypes(include=['int64', 'float64']).columns
for column in outlier_info.keys():
    if column in numeric_columns:  # Only operate on numeric columns
        Q1 = sampled_df_cleaned[column].quantile(0.25)
        Q3 = sampled_df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove rows with outliers
        sampled_df_cleaned = sampled_df_cleaned[
            (sampled_df_cleaned[column] >= lower_bound) & (sampled_df_cleaned[column] <= upper_bound)
        ]

# Print the processed data
print("\nData after removing outliers:")
print(sampled_df_cleaned.shape)
```

**Dataset Split and Model Selection**

For the tasks, evaluation metrics, and dataset described, I chose supervised learning, with the binary target variable indicating whether a student is depressed. The cleaned and preprocessed dataset (2,786 rows) was split into 70% for the training set and 30% for the testing set. The selected models included Decision Tree, Random Forest, SVM, and KNN.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix, classification_report

# Randomforest
y = sampled_df_cleaned['Depression']  # Target variable
X = sampled_df_cleaned.drop(columns=['Depression'])  # Features

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
y_train = y_train.astype(int)
```

The **Decision Tree** model is a tree-structured approach that recursively splits the dataset into subsets based on feature decision rules, ultimately classifying the data into categories [7]. Each node represents a feature's decision criterion, and the final leaves provide classification outcomes. This model is well-suited for the selected dataset (17 feature dimensions and 1 target variable) and the task of predicting student depression based on multiple features. Additionally, the Decision Tree model inherently provides feature importance, making it particularly valuable for analyzing the risk factors contributing to depression.

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree classifier
max_depth_dt = None  # You can adjust the maximum depth of the tree
decision_tree = DecisionTreeClassifier(max_depth=max_depth_dt, random_state=42)
decision_tree.fit(X_train, y_train)

# Make predictions on the test set
y_pred_dt = decision_tree.predict(X_test)
y_test_dt = y_test.astype(int)
y_pred_dt = y_pred_dt.astype(int)

# Confusion matrix
cm_dt = confusion_matrix(y_test_dt, y_pred_dt)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Depression', 'Depression'], yticklabels=['No Depression', 'Depression'])
plt.title('Confusion Matrix - Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Total Accuracy
total_accuracy_dt = accuracy_score(y_test_dt, y_pred_dt)

# Class-wise Metrics
class0_precision_dt = cm_dt[0, 0] / (cm_dt[0, 0] + cm_dt[1, 0])  # TN / (TN + FN)
class0_recall_dt = cm_dt[0, 0] / (cm_dt[0, 0] + cm_dt[0, 1])     # TN / (TN + FP)
class0_f1_dt = 2 * (class0_precision_dt * class0_recall_dt) / (class0_precision_dt + class0_recall_dt)

class1_precision_dt = cm_dt[1, 1] / (cm_dt[1, 1] + cm_dt[0, 1])  # TP / (TP + FP)
class1_recall_dt = cm_dt[1, 1] / (cm_dt[1, 1] + cm_dt[1, 0])     # TP / (TP + FN)
class1_f1_dt = 2 * (class1_precision_dt * class1_recall_dt) / (class1_precision_dt + class1_recall_dt)

# Print metrics
print("\nEvaluation Metrics - Decision Tree:")
print(f"Total Accuracy: {total_accuracy_dt:.2f}")
print(f"Class 0 (No Depression) - Precision: {class0_precision_dt:.2f}, Recall: {class0_recall_dt:.2f}, F1 Score: {class0_f1_dt:.2f}")
print(f"Class 1 (Depression) - Precision: {class1_precision_dt:.2f}, Recall: {class1_recall_dt:.2f}, F1 Score: {class1_f1_dt:.2f}")

# Feature Importances
feature_importances_dt = decision_tree.feature_importances_
feature_names_dt = X.columns
importance_df_dt = pd.DataFrame({'Feature': feature_names_dt, 'Importance': feature_importances_dt})
importance_df_dt = importance_df_dt.sort_values(by='Importance', ascending=False)  # Sort by importance

print("\nFeature Importances (sorted) - Decision Tree:")
print(importance_df_dt)

# Plot Feature Importances
plt.figure(figsize=(10, 6))
sns.barplot(x=importance_df_dt['Importance'], y=importance_df_dt['Feature'], palette="viridis")
plt.title("Feature Importances - Decision Tree")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()

# ROC Curve and AUC
roc_auc_dt = roc_auc_score(y_test_dt, decision_tree.predict_proba(X_test)[:, 1])
fpr_dt, tpr_dt, thresholds_dt = roc_curve(y_test_dt, decision_tree.predict_proba(X_test)[:, 1])

print(f"\nROC-AUC Score - Decision Tree: {roc_auc_dt:.2f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr_dt, tpr_dt, label=f"ROC Curve (AUC = {roc_auc_dt:.2f})", color='blue')
plt.plot([0, 1], [0, 1], 'r--', label='Random Chance')
plt.title('Receiver Operating Characteristic (ROC) Curve - Decision Tree')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from sklearn.svm import SVC

# SVM classifier
kernel_svm = 'linear'  # You can adjust the kernel ('linear', 'poly', 'rbf', 'sigmoid') as needed
C_svm = 1.0  # Regularization parameter
svm_model = SVC(kernel=kernel_svm, C=C_svm, probability=True, random_state=42)
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_svm = svm_model.predict(X_test)
y_test_svm = y_test.astype(int)
y_pred_svm = y_pred_svm.astype(int)

# Confusion matrix
cm_svm = confusion_matrix(y_test_svm, y_pred_svm)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Depression', 'Depression'], yticklabels=['No Depression', 'Depression'])
plt.title('Confusion Matrix - SVM')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Total Accuracy
total_accuracy_svm = accuracy_score(y_test_svm, y_pred_svm)

# Class-wise Metrics
class0_precision_svm = cm_svm[0, 0] / (cm_svm[0, 0] + cm_svm[1, 0])  # TN / (TN + FN)
class0_recall_svm = cm_svm[0, 0] / (cm_svm[0, 0] + cm_svm[0, 1])     # TN / (TN + FP)
class0_f1_svm = 2 * (class0_precision_svm * class0_recall_svm) / (class0_precision_svm + class0_recall_svm)

class1_precision_svm = cm_svm[1, 1] / (cm_svm[1, 1] + cm_svm[0, 1])  # TP / (TP + FP)
class1_recall_svm = cm_svm[1, 1] / (cm_svm[1, 1] + cm_svm[1, 0])     # TP / (TP + FN)
class1_f1_svm = 2 * (class1_precision_svm * class1_recall_svm) / (class1_precision_svm + class1_recall_svm)

# Print metrics
print("\nEvaluation Metrics - SVM:")
print(f"Total Accuracy: {total_accuracy_svm:.2f}")
print(f"Class 0 (No Depression) - Precision: {class0_precision_svm:.2f}, Recall: {class0_recall_svm:.2f}, F1 Score: {class0_f1_svm:.2f}")
print(f"Class 1 (Depression) - Precision: {class1_precision_svm:.2f}, Recall: {class1_recall_svm:.2f}, F1 Score: {class1_f1_svm:.2f}")

# Extract feature importance for linear SVM
feature_importances_svm = np.abs(svm_model.coef_).flatten()  # Absolute value of coefficients
feature_names_svm = X.columns

# Create a DataFrame for feature importance
importance_df_svm = pd.DataFrame({
    'Feature': feature_names_svm,
    'Importance': feature_importances_svm
}).sort_values(by='Importance', ascending=False)

# Print feature importance
print("\nFeature Importances (sorted) - SVM:")
print(importance_df_svm)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=importance_df_svm['Importance'], y=importance_df_svm['Feature'], palette="viridis")
plt.title("Feature Importances - SVM (Linear Kernel)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()


# ROC Curve and AUC
roc_auc_svm = roc_auc_score(y_test_svm, svm_model.decision_function(X_test))
fpr_svm, tpr_svm, thresholds_svm = roc_curve(y_test_svm, svm_model.decision_function(X_test))

print(f"\nROC-AUC Score - SVM: {roc_auc_svm:.2f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr_svm, tpr_svm, label=f"ROC Curve (AUC = {roc_auc_svm:.2f})", color='blue')
plt.plot([0, 1], [0, 1], 'r--', label='Random Chance')
plt.title('Receiver Operating Characteristic (ROC) Curve - SVM')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

The **K-NN** model classifies or predicts by calculating the distance between a new sample and the samples in the training data, identifying the \( K \) nearest neighbors, and using their categories to make decisions [5]. This approach is particularly suited for the binary classification task of determining whether a student is depressed based on multiple features. After training the K-NN model, depressed students may form distinct clustering patterns in the feature space, which the model leverages to predict and classify test samples. Although K-NN does not provide direct methods for feature importance analysis, importance can be indirectly assessed by comparing the baseline model's accuracy with the accuracy of a model trained after removing a specific feature, identifying the contribution of that feature to the model's performance.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# KNN classifier
n_neighbors_knn = 3  # Number of neighbors (you can adjust as needed)
knn_model = KNeighborsClassifier(n_neighbors=n_neighbors_knn)
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_knn = knn_model.predict(X_test)
y_test_knn = y_test.astype(int)
y_pred_knn = y_pred_knn.astype(int)

# Confusion matrix
cm_knn = confusion_matrix(y_test_knn, y_pred_knn)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['No Depression', 'Depression'], yticklabels=['No Depression', 'Depression'])
plt.title('Confusion Matrix - KNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Total Accuracy
total_accuracy_knn = accuracy_score(y_test_knn, y_pred_knn)

# Class-wise Metrics
class0_precision_knn = cm_knn[0, 0] / (cm_knn[0, 0] + cm_knn[1, 0])  # TN / (TN + FN)
class0_recall_knn = cm_knn[0, 0] / (cm_knn[0, 0] + cm_knn[0, 1])     # TN / (TN + FP)
class0_f1_knn = 2 * (class0_precision_knn * class0_recall_knn) / (class0_precision_knn + class0_recall_knn)

class1_precision_knn = cm_knn[1, 1] / (cm_knn[1, 1] + cm_knn[0, 1])  # TP / (TP + FP)
class1_recall_knn = cm_knn[1, 1] / (cm_knn[1, 1] + cm_knn[1, 0])     # TP / (TP + FN)
class1_f1_knn = 2 * (class1_precision_knn * class1_recall_knn) / (class1_precision_knn + class1_recall_knn)

# Print metrics
print("\nEvaluation Metrics - KNN:")
print(f"Total Accuracy: {total_accuracy_knn:.2f}")
print(f"Class 0 (No Depression) - Precision: {class0_precision_knn:.2f}, Recall: {class0_recall_knn:.2f}, F1 Score: {class0_f1_knn:.2f}")
print(f"Class 1 (Depression) - Precision: {class1_precision_knn:.2f}, Recall: {class1_recall_knn:.2f}, F1 Score: {class1_f1_knn:.2f}")

# Initialize a DataFrame to store feature importance
feature_importance_knn = []

# Baseline accuracy
baseline_accuracy = accuracy_score(y_test_knn, knn_model.predict(X_test))

# Evaluate importance by removing each feature
for feature in X.columns:
    X_train_reduced = X_train.drop(columns=[feature])
    X_test_reduced = X_test.drop(columns=[feature])

    # Train and evaluate a new KNN model
    knn_model_reduced = KNeighborsClassifier(n_neighbors=n_neighbors_knn)
    knn_model_reduced.fit(X_train_reduced, y_train)
    reduced_accuracy = accuracy_score(y_test_knn, knn_model_reduced.predict(X_test_reduced))

    # Importance as the drop in accuracy
    importance = baseline_accuracy - reduced_accuracy
    feature_importance_knn.append({'Feature': feature, 'Importance': importance})

# Convert to DataFrame and sort by importance
importance_df_knn = pd.DataFrame(feature_importance_knn).sort_values(by='Importance', ascending=False)

# Print feature importance
print("\nFeature Importances (sorted) - KNN:")
print(importance_df_knn)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=importance_df_knn['Importance'], y=importance_df_knn['Feature'], palette="viridis")
plt.title("Feature Importances - KNN")
plt.xlabel("Importance (Drop in Accuracy)")
plt.ylabel("Feature")
plt.show()


# ROC Curve and AUC
roc_auc_knn = roc_auc_score(y_test_knn, knn_model.predict_proba(X_test)[:, 1])
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test_knn, knn_model.predict_proba(X_test)[:, 1])

print(f"\nROC-AUC Score - KNN: {roc_auc_knn:.2f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr_knn, tpr_knn, label=f"ROC Curve (AUC = {roc_auc_knn:.2f})", color='blue')
plt.plot([0, 1], [0, 1], 'r--', label='Random Chance')
plt.title('Receiver Operating Characteristic (ROC) Curve - KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
